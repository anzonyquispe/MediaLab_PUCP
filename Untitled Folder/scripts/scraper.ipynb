```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException,TimeoutException
import os
import pandas as pd
import time

options = webdriver.ChromeOptions()

# Suponiendo monitor vertical de 1080px de alto x 1920 ancho
options.add_argument("--start-maximized")
options.add_argument("window-size=1080,1920")
options.add_argument("window-position=-1080,0")

url = "https://dilrmp.gov.in/faces/percent/rptComputerizationOfLandRecord.xhtml"

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

driver.get(url)


```

```python
filename = "step_one.xlsx"

if not os.path.exists(filename):

   print(f"Generating file {filename}")

   data = []
   states = driver.find_elements(By.XPATH, '//*[@id="myclrstateform:clrstatemanage_data"]/tr/td[2]')

   for state in states:

      anchor = state.find_element(By.TAG_NAME,"a")

      children_links  = anchor.get_attribute("href")

      state_name = anchor.text

      data.append({"STATE": state_name,"URL": children_links})

      df = pd.DataFrame(data)

       # Guardamos el archivo Excel
      df.index = range(1, len(df) + 1)
      df.to_excel(filename, index=True)
      driver.quit()

if os.path.exists(filename):
   print(f"The {filename} file already exists, let´s go step two.")
```

```python

filename2 = "step_two.xlsx"

if not os.path.exists(filename2):

   print(f"Generating file {filename2}")

   df = pd.read_excel("step_one.xlsx")

   data = []

   # for url in df["URL"]:

   for index, row in df.iterrows():
      state = row["STATE"]
      url = row["URL"]

      print(state,index)

      driver = webdriver.Chrome(service=service, options=options)

      driver.get(url)


      # *********Preparar el campo de extracción **************

      # Obtener el elemento(footer) que nos molesta
      elements = driver.find_elements(By.XPATH, '//*[@id="footer"]/div/div/div/div')

      if len(elements) > 0:
         elem = elements[0]

         # Ejecutar JS para eliminar ese elemento
         driver.execute_script("""
         var element = arguments[0];
         element.parentNode.removeChild(element);
         """, elem)
      else:
         print("El elemento molestoso ya no existe")

      # Modificar el elemento Options para  mostrar toda la data en un solo plano
      elem = driver.find_element(By.XPATH,'//*[@id="myform:compListTable_rppDD"]/option[3]')
      # Modificar html interno, texto y atributo
      driver.execute_script("arguments[0].innerHTML = '100'", elem)
      driver.execute_script("arguments[0].setAttribute('value', '100')", elem)
      driver.execute_script("arguments[0].click();", elem)
      elem.click()
      time.sleep(1.5)
      # Esperar a que la página termine de cargar
      WebDriverWait(driver, 10
                    ).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="myform"]')))

      # ********* fin de preparación del campo de extracción**************

      # ********* Extracción **************

      districts = driver.find_elements(By.XPATH, '//*[@id="myform:compListTable_data"]/tr/td[2]')

      for district in districts:

         anchor = district.find_element(By.XPATH, './/a')
         district_name = anchor.text
         children_links = anchor.get_attribute("href")

         data.append({"STATE": state, "DISTRICT":district_name, "URL": children_links})

      driver.quit()



   # ********* Fin de extracción **************

   df = pd.DataFrame(data)

   # Guardamos el archivo Excel
   df.to_excel(filename2, index=False)

if os.path.exists(filename2):
   print(f"The {filename2} file already exists, let´s go step three.")
```

```python

filename3 = "step_three.xlsx"

if not os.path.exists(filename3):

   print(f"Generating file {filename3}")

   df = pd.read_excel("step_two.xlsx")

   data = []

   for index, row in df.iterrows():
      state = row["STATE"]
      district = row["DISTRICT"]
      url = row["URL"]

      print(state, district, index)

      driver = webdriver.Chrome(service=service, options=options)

      driver.get(url)


      # *********Preparar el campo de extracción **************

      # Obtener el elemento(footer) que nos molesta
      elements = driver.find_elements(By.ID, 'footer')

      if len(elements) > 0:
         elem = elements[0]

         # Ejecutar JS para eliminar ese elemento
         driver.execute_script("""
         var element = arguments[0];
         element.parentNode.removeChild(element);
         """, elem)
      else:
         print("El elemento molestoso (footer) no existe")

      # Modificar el elemento Options para  mostrar toda la data en un solo plano
      elem = driver.find_element(By.XPATH,'//*[@id="myform:compListTable_rppDD"]/option[3]')
      # Modificar html interno, texto y atributo
      driver.execute_script("arguments[0].innerHTML = '10000'", elem)
      driver.execute_script("arguments[0].setAttribute('value', '10000')", elem)
      driver.execute_script("arguments[0].click();", elem)
      elem.click()
      time.sleep(1.4)
      # Esperar a que la página termine de cargar
      WebDriverWait(driver, 10
                    ).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="myform"]')))

      # ********* fin de preparación del campo de extracción**************

      # ********* Extracción **************

      tehsils = driver.find_elements(By.XPATH, '//*[@id="myform:compListTable_data"]/tr/td[2]')

      for tehsil in tehsils:

         anchor = tehsil.find_element(By.XPATH, './/a')
         tehsil_name = anchor.text
         children_links = anchor.get_attribute("href")

         data.append({"STATE": state, "DISTRICT":district, "TEHSIL":tehsil_name, "URL": children_links})

      driver.quit()



   # ********* Fin de extracción **************

   df = pd.DataFrame(data)

   # Guardamos el archivo Excel
   df.to_excel(filename3, index=False)

if os.path.exists(filename3):
   print(f"The {filename3} file already exists, let´s go step four.")
```
