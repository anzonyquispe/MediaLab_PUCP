```python
folder_save = r"/Users/ar8787/Dropbox/India_forest_land/B_RawData/scraping_dilrmp/subdist/original"

modified_dir = r"/Users/ar8787/Dropbox/India_forest_land/B_RawData/scraping_dilrmp/subdist/modified"

# 
service = Service(executable_path= r"/Users/ar8787/Downloads/chromedriver-mac-arm64/chromedriver")
options = webdriver.ChromeOptions()

# Set the default download directory
download_dir = folder_save
prefs = {"download.default_directory" : download_dir}
options.add_experimental_option("prefs", prefs)
driver = webdriver.Chrome(service=service, options=options)

url = "https://dilrmp.gov.in/faces/rptdistrictwisephysical/rptComputerizationOfLandRecord.xhtml?statecode=2"
driver.get(url) #navigate to the page
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'container')))

dist_name_list = []
dist_code_list = []
dist_link_list = []
click_next = True

while click_next:
    
    # WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'container')))
    time.sleep(20)
    table_dist = driver.find_element( By.XPATH, f'//*[@id="myform:compListTable_data"]')
    rows = table_dist.find_elements(By.TAG_NAME, 'tr')
    total_dist_rows = len(rows)

    
    for dist_nrow in np.arange(1, total_dist_rows + 1 ):
        dist_ele = driver.find_element( By.XPATH, f'//*[@id="myform:compListTable_data"]/tr[{dist_nrow}]/td[2]/a')
        dist_link = dist_ele.get_attribute('href')
        dist_code = dist_link.split('=')[-1]
        dist_name = dist_ele.text

        dist_name_list.append( dist_name )
        dist_code_list.append( dist_code )
        dist_link_list.append( dist_link )
        
    next_dist_but = driver.find_element( By.XPATH,'//*[@id="myform:compListTable_paginator_bottom"]/span[5]')
    class_dist_value = next_dist_but.get_attribute("class")

    if "ui-state-disabled" in class_dist_value:
        print("The element contains the 'ui-state-disabled' class. There are not more pages.")
        break
    else:
        print("The element does not contain the 'ui-state-disabled' class.")
        next_dist_but.click()

df1 = pd.DataFrame({"dist_name": dist_name_list, "dist_code" : dist_code_list, "dist_link" : dist_link_list  })

df1.to_csv(r"/Users/ar8787/Dropbox/India_forest_land/B_RawData/scraping_dilrmp/dist/state_2.csv", index = False )

subdist_name_list = []
subdist_code_list = []
subdist_link_list = []
dist_name_list2 = []
dist_code_list2 = []
dist_link_list2 = []

for dist_name, dist_code, dist_link in zip(dist_name_list, dist_code_list, dist_link_list):

    driver.get( dist_link )
    driver.execute_script("document.body.style.zoom='15%'")
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'container')))
    
    click_next = True
    while click_next:

        # WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'container')))
        time.sleep(20)
        table_subdist = driver.find_element( By.XPATH, f'//*[@id="myform:compListTable_data"]')
        rows = table_subdist.find_elements(By.TAG_NAME, 'tr')
        total_subdist_rows = len(rows)


        for subdist_nrow in np.arange(1, total_subdist_rows + 1 ):
            subdist_ele = driver.find_element( By.XPATH, f'//*[@id="myform:compListTable_data"]/tr[{subdist_nrow}]/td[2]/a')
            subdist_link = subdist_ele.get_attribute('href')
            subdist_code = subdist_link.split('=')[-1]
            subdist_name = subdist_ele.text

            subdist_name_list.append( subdist_name )
            subdist_code_list.append( subdist_code )
            subdist_link_list.append( subdist_link )
            
            dist_name_list2.append( dist_name )
            dist_code_list2.append( dist_code )
            dist_link_list2.append( dist_link )

        next_subdist_but = driver.find_element( By.XPATH,'//*[@id="myform:compListTable_paginator_bottom"]/span[5]')
        class_subdist_value = next_subdist_but.get_attribute("class")

        if "ui-state-disabled" in class_subdist_value:
            print("The element contains the 'ui-state-disabled' class. There are not more pages.")
            break
        else:
            print("The element does not contain the 'ui-state-disabled' class.")
            driver.execute_script("arguments[0].click();", next_subdist_but)

df2 = pd.DataFrame({"dist_name": dist_name_list2, "dist_code" : dist_code_list2, "dist_link" : dist_link_list2, 
                   "subdist_name": subdist_name_list, "subdist_code" : subdist_code_list, "subdist_link" : subdist_link_list})

df2.to_csv(r"/Users/ar8787/Dropbox/India_forest_land/B_RawData/scraping_dilrmp/subdist/state_2_subdist_links.csv", index = False )
```

```python
for i in range(36, 40):
    df2 = pd.read_csv(f"/Users/ar8787/Dropbox/India_forest_land/B_RawData/scraping_dilrmp/subdist/state_{i}_subdist_links.csv" )
    print(df2.shape)

    dist_name_list = df2.dist_name.tolist()
    dist_code_list = df2.dist_code.tolist()
    dist_link_list = df2.dist_link.tolist()
    subdist_name_list = df2.subdist_name.tolist()
    subdist_code_list = df2.subdist_code.tolist()
    subdist_link_list = df2.subdist_link.tolist()

    folder_save = fr"/Users/ar8787/Dropbox/India_forest_land/B_RawData/scraping_dilrmp/subdist/state{i}/original"

    modified_dir = fr"/Users/ar8787/Dropbox/India_forest_land/B_RawData/scraping_dilrmp/subdist/state{i}/modified"

    # 
    service = Service(executable_path= r"/Users/ar8787/Downloads/chromedriver-mac-arm64/chromedriver")
    options = webdriver.ChromeOptions()

    # Set the default download directory
    download_dir = folder_save
    prefs = {"download.default_directory" : download_dir}
    options.add_experimental_option("prefs", prefs)
    driver = webdriver.Chrome(service=service, options=options)

    zip_all_lists = zip( dist_name_list, dist_code_list, dist_link_list, 
                        subdist_name_list, subdist_code_list, subdist_link_list )

    for dist_name, dist_code, dist_link, subdist_name, subdist_code, subdist_link in zip_all_lists:

        url = subdist_link
        # Maximum number of attempts
        max_attempts = 5
        attempts = 0

        while attempts < max_attempts:
            try:
                # Navigate to your page
                driver.get(url)

                # Wait for the element to be present
                element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.ID, 'container'))
                )

                # If the element is found, break out of the loop
                if element:
                    print("Element found")
                    break
            except Exception as e:
                print(f"Attempt {attempts + 1} failed: {e}")
                attempts += 1
                if attempts == max_attempts:
                    print("Max attempts reached, element not found")
                    break
                else:
                    time.sleep(300)
                    print("Retrying...")

        # check the number of files
        files_before = os.listdir(download_dir)

        # download the file
        excel_ele = driver.find_element(By.XPATH, "/html/body/div[2]/div/form/table/tbody/tr[7]/td/input[1]")
        excel_ele.click()

        # Wait for a new file to appear in the folder
        new_file = None
        timeout = 60000  # Set a timeout for 5 minutes
        check_interval = 5  # Check every 5 seconds

        start_time = time.time()
        while (time.time() - start_time) < timeout:
            # Get the list of files in the directory
            files_after = os.listdir(download_dir)
            # Filter out files with ".crdownload" extension
            non_crdownload_files_after = [file for file in files_after if file.endswith('.xls') and not 'crdownload' in file]
            # Check for the difference
            new_files = set(non_crdownload_files_after) - set(files_before)
            if new_files:
                new_file = new_files.pop()
                print(f"New file detected: {new_file}")
                time.sleep(5)
                break
            time.sleep(check_interval)

        # Get the most recent file in the download directory
        files = os.listdir(download_dir)
        paths = [os.path.join(download_dir, basename) for basename in files if not basename.endswith('.DS_Store') and not 'crdownload' in basename ]
        latest_file = max(paths, key=os.path.getctime)

        # New file name (ensure it ends with .xlsx if it's an Excel file)
        newfilename = f"{int(dist_code)}_{int(subdist_code)}.xls"
        print(dist_name, "-", subdist_name)
        new_file_name = os.path.join(download_dir, newfilename)

        # Rename the file
        os.rename(latest_file, new_file_name)

        # Adding state name and dist name
        df = pd.read_excel(new_file_name)
        df['dist_name'] = dist_name
        df['dist_code'] = dist_code
        df['dist_link'] = dist_link
        df['subdist_name'] = subdist_name
        df['subdist_code'] = subdist_code
        df['subdist_link'] = subdist_link


        # Saving new file
        modi_newname = newfilename.replace('xls', 'xlsx')
        modi_filepath = os.path.join(modified_dir, modi_newname)
        df.to_excel( modi_filepath, index = False )
```
