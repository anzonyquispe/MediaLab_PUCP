```python
#task 1 We decided to show some descriptive statistics from "Operating revenue (Turnover)\nth LCU 2023" to "Operating revenue (Turnover)\nth USD 2021"
import pandas as pd

# Define file path
file_path = "../../_data/bogota_firms.xlsx"

# Read the Excel file into a DataFrame
dfr = pd.read_excel(file_path, sheet_name="Results")

# Display the DataFrame
dfr
```

```python
# Specify the columns for descriptive statistics
start_column = "Operating revenue (Turnover)\nth LCU 2023"
end_column = "Operating revenue (Turnover)\nth USD 2021"
selected_columns = dfr.loc[:, start_column:end_column]

# Show descriptive statistics for the selected columns
statistics = selected_columns.describe()
print(statistics)
```

```python
#task 2
# Use the drop function to remove columns containing the specified label
dfr = dfr.drop(dfr.filter(like='Operating revenue (Turnover)\nth USD').columns, axis=1)
dfr
```

```python
# Melt the DataFrame to convert it from wide to long format
melted_dfr = pd.melt(dfr, id_vars=['Company name Latin alphabet', 'Last avail. year'], 
                    var_name='Operating year', value_name='Operating revenue (Turnover)')

# Filter out rows where 'Operating revenue (Turnover)' is not available
melted_dfr = melted_dfr[melted_dfr['Operating revenue (Turnover)'] != 'n.a.']

# Convert 'Operating year' column to numeric type
melted_dfr['Operating year'] = pd.to_numeric(melted_dfr['Operating year'].str.extract('(\d+)')[0])

# Display the resulting DataFrame
print(melted_dfr)
```

```python
dfr_long = pd.melt(dfr, id_vars=['Last avail. year'], var_name='variable', value_name='valor')

# Show the resulting DataFrame
print(dfr_long)
```

```python
#Task 5
# create a funtion
def calling_data(firm_name, start_year, end_year=None):
    
    # Convert the company name to a list
    firm_names_list = [firm_name] if isinstance(firm_name, str) else firm_name
    
    # Filter data by company name using isin
    filtered_data = dfr[dfr['Company Name'].isin(firm_names_list)]
    
    # Filter data by year range usion Last Year
    filtered_data = filtered_data[(filtered_data['Last Year'] >= start_year) & (filtered_data['Last Year'] <= end_year)] if end_year else filtered_data
    
    # Select required columns, using f for reducing the text
    selected_columns = ['Company Name', 'City', 'NACE Code', 'Last Year'] + [f'Revenue {year} LCU' for year in range(2023, start_year - 1, -1)]

    # Create a new dataframe with the selected columns
    result_df = filtered_data[selected_columns]

    # Check if result_df is empty
    if result_df.empty:
        print(f"No hay datos para la empresa {firm_name} en el rango de años especificado.")
        return None, None

    # Get the names of revenue variables
    revenue_columns = [f'Revenue {year} LCU' for year in range(2023, start_year - 1, -1)]

    # Create a dictionary mapping variable names to values
    variables_dict = {column: result_df.iloc[0][column] for column in revenue_columns}

    return result_df, variables_dict
```

```python
# Expample
firm_name = 'CONTACTO ESTRATEGICO OUTSOURCING CONTABLE S A S'
start_year = 2016
end_year = 2022

result_df, variables_dict = calling_data(firm_name, start_year, end_year)
if result_df is not None:
    print(result_df)
    print(variables_dict)
```
