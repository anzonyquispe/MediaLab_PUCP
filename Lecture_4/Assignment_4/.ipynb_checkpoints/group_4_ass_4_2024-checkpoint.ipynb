```python
#task 1 We decided to show some descriptive statistics from "Operating revenue (Turnover)\nth LCU 2023" to "Operating revenue (Turnover)\nth USD 2021"
import pandas as pd

# Define file path
file_path = "../../_data/bogota_firms.xlsx"

# Read the Excel file into a DataFrame
dfr = pd.read_excel(file_path, sheet_name="Results")

# Display the DataFrame
dfr
```

```python
# Specify the columns for descriptive statistics
start_column = "Operating revenue (Turnover)\nth LCU 2023"
end_column = "Operating revenue (Turnover)\nth USD 2021"
selected_columns = dfr.loc[:, start_column:end_column]

# Show descriptive statistics for the selected columns
statistics = selected_columns.describe()
print(statistics)
```

```python
#task 2
# Use the drop function to remove columns containing the specified label
dfr = dfr.drop(dfr.filter(like='Operating revenue (Turnover)\nth USD').columns, axis=1)
dfr
```

```python
# Melt the DataFrame to convert it from wide to long format
melted_dfr = pd.melt(dfr, id_vars=['Company name Latin alphabet', 'Last avail. year'], 
                    var_name='Operating year', value_name='Operating revenue (Turnover)')

# Filter out rows where 'Operating revenue (Turnover)' is not available
melted_dfr = melted_dfr[melted_dfr['Operating revenue (Turnover)'] != 'n.a.']

# Convert 'Operating year' column to numeric type
melted_dfr['Operating year'] = pd.to_numeric(melted_dfr['Operating year'].str.extract('(\d+)')[0])

# Display the resulting DataFrame
print(melted_dfr)
```

```python
dfr_long = pd.melt(dfr, id_vars=['Last avail. year'], var_name='variable', value_name='valor')

# Show the resulting DataFrame
print(dfr_long)
```
