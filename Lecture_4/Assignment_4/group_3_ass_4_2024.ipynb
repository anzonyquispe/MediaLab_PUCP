### Second exercise

```python
#Reshape the dataset from wide to long at year level. Drop columns if they are in Dollars. 
#Keep columns from Operating revenue (Turnover) th LCU 2023 to the end, and the name of the firm.
```

```python
# Select specific columns by their positions
selected_columns = bogota_firms_sheet2.columns[[1] + list(range(6, len(bogota_firms_sheet2.columns)))]

# Create a new DataFrame with the selected columns
selected_df = bogota_firms_sheet2[selected_columns]

print("\nSelected DataFrame:")
display(selected_df)
```

```python
# Reshape the DataFrame to long format
selected_df_reshaped = selected_df.set_index(  [ 'Company name Latin alphabet' ,'Last avail. year']  ).stack().reset_index().rename( {"REVENUE" : "VARIABLES", 
                                                                                0 : "VALUES" }, axis = 1 )
selected_df_reshaped.head()

print("\nSorted Long-format DataFrame:")
display(selected_df_reshaped)
```

```python

# Filter to drop rows where 'Variable' contains 'USD'
selected_df_reshaped_filtered = selected_df_reshaped[~selected_df_reshaped['level_2'].str.contains('USD')]

print("\nFiltered Long-format DataFrame:")
selected_df_reshaped_filtered_reset = selected_df_reshaped_filtered.reset_index(drop=True)

print("\nDataFrame with Reset Index:")
display(selected_df_reshaped_filtered_reset)
```

### Third exercise

```python
#Renaming the columns using ".rename"
selected_df_reshaped_filtered_reset.rename(columns={
    'Company name Latin alphabet': 'Company name (L.A.)',
    'Last avail. year': 'Last year',
    'level_2': 'Operating Revenue',
    'VALUES': 'Values (from O.R.)'
}, inplace=True)

print("\nDataFrame with renaming columns:")
display(selected_df_reshaped_filtered_reset)
```

### Fourth exercise

```python
#Generating a dictionary with the definition of all the columns in the  dataset
definitions_columns = {
    'Company name (L.A.)': "Company name in Latin alphabet",
    'Last year': "Latest year available for data",
    'Operating Revenue': "Operating revenue for each year",
    'Values (from O.R.)': "Values established for each operating revenue",
}

#Showing the definition for each column
print("\nColumns definition:\n")
for column, definition in definitions_columns.items():
    print(f"{column}: {definition}")
```

### Fifth exercise

```python
import pandas as pd
import re

def extract_year(year_str):
    match = re.search(r'\b(\d{4})\b', year_str)
    return int(match.group(1)) if match else None

def calling_data(firm_name, start_year=None, end_year=None):
    # Extract years from 'Operating Revenue' column and 'Last year' column
    operating_years = selected_df_reshaped_filtered_reset['Operating Revenue'].apply(extract_year)
    last_years = selected_df_reshaped_filtered_reset['Last year']

    # Determine start_year if not provided
    if start_year is None:
        start_year = min(operating_years)

    # Determine end_year if not provided
    if end_year is None:
        end_year = max(last_years)

    # Extract years from 'Operating Revenue' column and create a new 'OperatingYear' column
    selected_df_reshaped_filtered_reset['OperatingYear'] = operating_years

    # Filter DataFrame based on input criteria
    selected_df_reshaped_filtered_reset_filtered = selected_df_reshaped_filtered_reset[(selected_df_reshaped_filtered_reset['Company name (L.A.)'].isin(firm_name)) & (selected_df_reshaped_filtered_reset['OperatingYear'] >= start_year) & (selected_df_reshaped_filtered_reset['Last year'] <= end_year)].reset_index(drop=True)

       # Return a tuple with the filtered DataFrame and column definitions
    return selected_df_reshaped_filtered_reset_filtered, definitions_columns

# Example usage:
firm_name = selected_df_reshaped_filtered_reset['Company name (L.A.)'].unique()
start_year = 2011

result_df, definitions_columns = calling_data(firm_name, start_year)
print("Filtered DataFrame:")
display(result_df)
```
