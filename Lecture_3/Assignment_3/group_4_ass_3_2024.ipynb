Get the min, max, sd, n_obs, n_missing for the following columns **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)**. We want a dataframe with the following columns **Variables, Min, Max, Mean, N_obs, N_missing** and sort by the number of missing rows. **Hint: Use `describe` and `pivot` methods**

```python
import pandas as pd

# Assuming endes_2019 is your DataFrame
# Extract the specified columns
selected_columns = ['V201', 'V613', 'V715', 'V511']
df_selected = endes_2019[selected_columns]

# Compute summary statistics using describe
summary_stats = df_selected.describe().transpose()

# Count missing values
n_missing = df_selected.isnull().sum()

# Create a DataFrame with summary statistics and count of missing values
result_df = pd.DataFrame({
    'Variables': summary_stats.index,
    'Min': summary_stats['min'],
    'Max': summary_stats['max'],
    'Mean': summary_stats['mean'],
    'N_obs': summary_stats['count'],
    'N_missing': n_missing
})

# Sort by the number of missing rows
result_df = result_df.sort_values(by='N_missing', ascending=False)

# Display the result
print(result_df)
```

7. Use `endes_2019` data to generate a new object named `mean_key_vars` to find the mean of **total children ever born (V201)**, **Ideal number of children (V613)**, **Husbands education-single yrs (V715)**, and **Age at first marriage (V511)** by year and department **(V024)**. Name these columns as **mean_total_children, mean_ideal_children, mean_hb_yr_educ and mean_first_marriage**, respectively. **Hint: Use groupby and [this link](https://stackoverflow.com/questions/40901770/is-there-a-simple-way-to-change-a-column-of-yes-no-to-1-0-in-a-pandas-dataframe).**

```python
import pandas as pd

# Assuming endes_2019 is your DataFrame

# Convert columns with binary answers from strings to integers
binary_columns = ['V201', 'V613', 'V715', 'V511']
endes_2019[binary_columns] = endes_2019[binary_columns].replace({'No': 0, 'Si': 1})

# Group by 'year' and 'department (V024)' and calculate the mean for each variable
grouped_mean = endes_2019.groupby(['year', 'V024']).agg({
    'V201': 'mean',
    'V613': 'mean',
    'V715': 'mean',
    'V511': 'mean'
}).reset_index()

# Rename the columns
grouped_mean.columns = ['year', 'V024', 'mean_total_children', 'mean_ideal_children', 'mean_hb_yr_educ', 'mean_first_marriage']

# Create the new DataFrame mean_key_vars
mean_key_vars = grouped_mean

# Display the result
print(mean_key_vars)
```

8. Reshape `mean_key_vars` from wide to long. We want a dataframe with three columns **dpto, variables, values**. Name this object as `reshape_mean_key_vars`. **Hint: Use melt method**.

```python
# Use 'melt' to transform the DataFrame to long format
reshape_mean_key_vars = pd.melt(mean_key_vars, id_vars=['year', 'V024'], 
                                var_name='variables', value_name='values')
#id_vars are the columns that will remain unchanged
#var_name is the name of the new column that will contain the variable names
#value_name is the name of the new column that will contain the values associated with those variables

# Rename the columns
reshape_mean_key_vars.columns = ['year', 'dpto', 'variables', 'values']

# result
print(reshape_mean_key_vars)
```

9. Replicate your answers for questions 7 and 8, but in one line of code. Make it the most simple as possible. **NO HINT**

```python
# .replace({'No': 0, 'Yes': 1}): Replaces the values 'No' with 0 and 'Yes' with 1
# .groupby(['year', 'V024']): Groups the resulting DataFrame by the 'year' and 'V024' (department) columns.
# .agg({'V201': 'mean', 'V613': 'mean', 'V715': 'mean', 'V511': 'mean'}): Calculates the mean for the specified columns ('V201', 'V613', 'V715', 'V511') within each group.
# .reset_index(): Resets the index after clustering to get a flat DataFrame.
# .melt: Applies the melt function to convert the DataFrame from wide to long format
# .rename(columns={'V024': 'dept'}): Renames column 'V024' to 'dept', according to the specifications.

reshape_mean_key_vars = pd.melt(
    endes_2019.replace({'No': 0, 'Si': 1}).groupby(['year', 'V024']).agg({'V201': 'mean', 'V613': 'mean', 'V715': 'mean', 'V511': 'mean'}).reset_index(),
    id_vars=['year', 'V024'],
    var_name='variables',
    value_name='values'
).rename(columns={'V024': 'dpto'})
reshape_mean_key_vars
```
